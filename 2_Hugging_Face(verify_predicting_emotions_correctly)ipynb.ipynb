**Load the pretrained emotion model**

```python
model_id = "j-hartmann/emotion-english-distilroberta-base"
clf = pipeline("text-classification", model=model_id, return_all_scores=False)
```

**Batch prediction example**

```python
texts = {
    "anger": "I can’t believe they did that to me — I’m absolutely furious!",
    "disgust": "That was so gross I couldn’t even look at it.",
    "fear": "I’m really scared something bad is going to happen.",
    "joy": "I’m so happy right now — everything is going perfectly!",
    "neutral": "It’s fine, nothing good or bad about it, just okay.",
    "sadness": "I feel so alone and heartbroken today.",
    "surprise": "I didn’t expect that at all — what a shocking moment!"
}
```

```python
for emotion, text in texts.items():
    pred = clf(text)[0]
    print(f"\nEXPECTED: {emotion.upper()}")
    print("TEXT:", text)
    print("MODEL PRED:", pred)
```
